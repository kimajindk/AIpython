```python
#11. 실습 - 딕셔너리로 그래프 그리기
notebook = {'g○○':150,'t○○':100,'m○○':160,'s○○':170}

plt.figure(figsize=(5,3))

# 막대 그래프의 x축은 딕셔너리의 키(제품이름)를 사용하고, 막대 그래프의 Y축은 딕셔너리의 값(가격)을 사용하고, 각 막대의 너비를 0.7 단위로 설정
plt.bar(notebook.keys(),notebook.values(), width = 0.7)

plt.xlabel('product name')
plt.ylabel('price')
plt.show()
```

```python
#12. 실습 - 딕셔너리로 그래프 그리기 + 점에 주석 추가
notebook = {'g○○':150,'t○○':100,'m○○':160,'s○○':170}

plt.figure(figsize=(5,3))
plt.scatter(notebook.keys(),notebook.values())
for i in range(len(notebook)):
  plt.annotate(list(notebook.values())[i], (list(notebook)[i],list(notebook.values())[i]),xytext = (5,-15), textcoords='offset points')
plt.xlabel('product name')
plt.ylabel('price')
plt.show()
```

```python
#13. 실습
year = [2000, 2005, 2010, 2015, 2020]
single_person = [15.5, 20.0, 23.9, 27.2, 31.7]
baby_born = [640089, 438707, 470171, 438420, 272337]

plt.figure(figsize=(5,3))
plt.scatter(single_person, baby_born, c = 'b')
 # 1인 가구 비율(single_person)과 출생아 수(baby_born) 사이의 관계를 나타내는 산점도를 그립니다.
for i in range(len(year)):
  plt.annotate(year[i],xy=(single_person[i], baby_born[i]),xytext=(5,-10),textcoords='offset points') #각 연도에 해당하는 데이터 지점 옆에 해당 연도를 표시하는 주석을 추가
plt.xlim(14,35)
plt.ylim(230000,700000)
plt.xlabel('single-person household ratio')
plt.ylabel('the number of babies born')
plt.show()
```

```python
import matplotlib.pyplot as plt

exercise ={'climbing':0,'soccer':0,'swimming':0,'tennis':0}

print('좋아하는 운동을 입력하세요.')
print('복수 개의 운동을 입력하려면 콤마로 구분하세요.')

while True:
  s = input('climbing,soccer,swimming,tennis,종료: ')
  if s=='종료':
    break
  for a in s.split(','):
    if a in exercise:
      # 해당 운동의 카운트를 1 증가시키는 연산
    else:
      print(f'{a}는 등록되지 않은 운동입니다.')

plt.figure(figsize=(5.5,3))
plt.bar(exercise.keys(), exercise.values(), color='g', width=0.6)
#plt.xticks(rotation='vertical')
plt.xlabel('exercise')
plt.ylabel('frequency')
plt.show()
```

```python
#연습문제 1 : 다음 데이터를 이용하여 1년 동안의 월별 평균 강수량을 꺾은선 그래프로 그려보세요.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
pre = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize = (12,6)) #그래프 사이즈 설정
plt.ylim(0,400) #최대, 최소 설정
plt.plot(months, pre, 'md-.', label = 'precipitation') #x, y, 선 색상,  마커 모양, 선 모양 (color = 'b', marker = 'x', linestyle =':')
for i in range(len(months)):
  plt.annotate(pre[i], (months[i],pre[i]),xytext = (2,5), textcoords='offset points') #좌표 표시
plt.title('Monthly Precipitation')
plt.xlabel('Month')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```
