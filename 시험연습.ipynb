```python
#10. 실습
year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
jpn = [523, 529, 536, 532, 527]

plt.figure(figsize = (5, 3))
# 그래프가 가로와 세로 크기를 각각 5인치와 3인치 크기 설정
# year 데이터를 x축으로, kor 데이터를 y축으로 하여 빨간색 점선으로 선 그래프를 그리고 범례에 'Korea'라고 표시
plt.plot(year, kor, 'r:', label ='Korea')
plt.plot(year, usa, 'b:', ) #범례에 'Korea'라고 표시
plt.plot(year, jpn, 'g-', label='Japan')
plt.ylim(400, 600)  # y축 범위 설정
plt.legend()# 그래프에 범례를 추가
plt.title('Mathematics performance')
plt.xlabel('year')
plt.ylabel('score')
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
pre = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize = (12,6)) #그래프 사이즈 설정
plt.ylim(0,400) #최대, 최소 설정
plt.plot(months, pre, 'md-.', label = 'precipitation') #x, y, 선 색상,  마커 모양, 선 모양 (color = 'b', marker = 'x', linestyle =':')
for i in range(len(months)):
  plt.annotate(pre[i], (months[i],pre[i]),xytext = (2,5), textcoords='offset points') #좌표 표시
plt.title('Monthly Precipitation')
plt.xlabel('Month')
plt.ylabel('Precipitation (mm)')
#for i, value in enumerate(pre):
#  plt.text(i, value+10, str(value), ha = 'center', va = 'bottom')
plt.legend()
plt.show()
```
